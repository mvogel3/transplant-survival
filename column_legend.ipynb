```python
import pandas as pd
```

```python
#In order to prevent having the first two rows of our raw excel data be the column names we are dropping
#the second row, although more readable, it sometimes vary slightly across Excel workbooks.
#To map the more descriptive "column name" to the acronym but better for data storing names, we are creating this
#legend.

#import an excel file from the 2012 to 2014 excel data to get the two column names
filepath = 'kidney/kidney_2012_07.XLS'
sheet_name = 'Table 7 - Deceased'
df = pd.read_excel(filepath, sheet_name)
```

```python
df.head()
```

```python
#drop everything except the column names and first row
df_col = df.iloc[[0]]
df_col.head()
```

```python
df_col.columns
```

```python
#transpose the columns into rows
df_col_transp = df_col.T
df_col_transp.head(30)
```

```python
#send legend to EXCEL
#df_col_transp.to_excel('output/LEGEND_Table_7_Deceased.xlsx')
df_col.to_excel('output/COLUMNS_Table_7_Deceased.xlsx')
```
