```python
import pandas as pd
```

```python
# Read the Excel file with multiple tabs
excel_file = pd.ExcelFile("Resources/HR-1207.xlsx")
excel_file2 = pd.ExcelFile("Resources/HR-1301.xlsx")
excel_file3 = pd.ExcelFile("Resources/HR-1307.xlsx")
excel_file4 = pd.ExcelFile("Resources/HR-1401.xlsx")
excel_file5 = pd.ExcelFile("Resources/HR-1406.xlsx")
excel_file6 = pd.ExcelFile("Resources/HR-1412.xlsx")
```

```python
# Get the list of sheet names
sheet_names = excel_file.sheet_names
```

```python
# Read data from a specific sheet by name
HR_1207 = excel_file.parse(sheet_names[0])
HR_1207.head()
```

```python
HR_1301 = excel_file2.parse(sheet_names[0])
HR_1307 = excel_file3.parse(sheet_names[0])
HR_1401 = excel_file4.parse(sheet_names[0])
HR_1406 = excel_file5.parse(sheet_names[0])
HR_1412 = excel_file6.parse(sheet_names[0])

HR_1412.head()
```

```python
HR_1207.columns
```

```python
HR_1412.columns
```

```python
HR_up_to_2014 = pd.concat([HR_1207, HR_1301, HR_1307, HR_1401, HR_1406, HR_1412])
HR_up_to_2014.head()
```

```python
HR_up_to_2014.rename(columns = {"Unnamed: 0":"ENTIRE_NAME"}, inplace=True)
HR_up_to_2014
```

```python
df = HR_up_to_2014.copy()
df = df.drop(index=[row for row in df.index if 'Comment' in df.loc[row].values])
```
