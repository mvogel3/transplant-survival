```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv('top-20-hospitals.csv')
data_df
```

```python
data_df_1 = data_df.loc[data_df["TYPE"]=="R"]
data_df_1
```

```python
data_df_2 = data_df.loc[data_df["TYPE"]=="C"]
data_df_2
```

```python
import folium
import numpy as np
from geopy.distance import geodesic

m = folium.Map([40.7, -94.5], zoom_start=4)


for index, row in data_df_1.iterrows():
    popContent = (row[0] + "<hr>" + str(row[3]))
    iframe = folium.IFrame(popContent)
    popup1 = folium.Popup(iframe,
                         min_width=200,
                         max_width=200)
    folium.CircleMarker([row[1], row[2]], popup=popup1,color='purple',
    fill_opacity=1.0,
    fill_color='purple').add_to(m)
    
for index, row in data_df_2.iterrows():
    popContent = (row[0] + "<hr>" + str(row[3]))
    iframe = folium.IFrame(popContent)
    popup1 = folium.Popup(iframe,
                         min_width=200,
                         max_width=200)
    folium.CircleMarker([row[1], row[2]], popup=popup1,color='pink',
    fill_opacity=1.0,
    fill_color='pink').add_to(m)
    
m
```

```python
file_path = 'm.html'
m.save(file_path)
```
