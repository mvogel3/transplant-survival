```python
sheet_name = 'Blood_Type'

abo_df = pd.read_excel('src/Transplant___Recipient_ABO_by_Transplant_Year,_Recipient_Age.xlsx', sheet_name=sheet_name)
```

```python
abo_df.head()
```

```python
#drop all the overall rows

certain_value = 'Overall'

abo_df.drop(abo_df[abo_df['Transplant Year'] == certain_value].index, inplace=True)
abo_df.head()
```

```python
#remove all recipient ages values
certain_value = 'All Recipient Ages'
abo_df.drop(abo_df[abo_df['Age Group'] == certain_value].index, inplace=True)
abo_df.head()
```

```python
#remove Unknown values
certain_value = 'Unknown'
abo_df.drop(abo_df[abo_df['Age Group'] == certain_value].index, inplace=True)
abo_df.head()
```

```python
#check the number of unique rows
unique = abo_df['Transplant Year'].value_counts()
print(len(unique))
```

```python
abo_df.to_csv('output/abo_df.csv')
```
