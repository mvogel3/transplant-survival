```python
import pandas as pd
import numpy as np
```

```python
kidney_df=pd.read_excel(io="merged/15_Year.xlsx", sheet_name="15_Year")
kidney_df
```

```python
kidney_df.columns
```

```python
for column in kidney_df.columns:
    if column not in ['Transplant Year', 'Age Group']:
        kidney_df[column] = pd.to_numeric(kidney_df[column], errors='coerce')
        
kidney_df.fillna(0,inplace=True)
kidney_df
```

```python
kidney_df.to_csv("merged/15_Year_cleaned.csv")
```
