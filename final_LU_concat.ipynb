```python
# Importing dependencies

import pandas as pd
import numpy as np
import datetime as dt
import re
import calendar
```

```python
# Concatenating the final files

# the remaining files have 195 columns and this loop adds 2 more

sheet_names = 'Figures C1-C2 - Deceased'
files = ['LU-1911.xls', 'LU-2006.xls', 'LU-2011.xls', 'LU-2105.xls', 'LU-2111.xls', 'LU-2205.xls', 'LU-2211.xls']

concat=False;
for file in files:
    year_month = re.sub("[^\d]", "", file)
    year = "20"+year_month[0:2]
    month = int(year_month[2:4])
    
    end_df = pd.read_excel(io='Resources_LU/'+file, sheet_name=sheet_names)
    
    end_df['YEAR']=year
    end_df['MONTH']=month
    
    if(concat==True):
        last_df = pd.concat([last_df, end_df], axis=0)
    else:
        last_df=end_df
        concat=True

last_df
```

```python
# appending donor status
final_df = last_df.copy()
final_df["Donor Status"] = "Deceased"
final_df.head()
```

```python
# making COHORT_DATE column

final_df['COHORT_DATE'] = final_df['YEAR'].astype(str) + '-' + final_df['MONTH'].astype(str)
```

```python
# editing COHORT_DATE format

final_df['COHORT_DATE'] = pd.to_datetime(final_df['COHORT_DATE']).dt.strftime('%Y-%m')
final_df.head()
```

```python
# removing the column name descriptor rows from the concatenated df

clean_final = final_df.copy()
clean_final = clean_final.drop(index=[row for row in clean_final.index if 'Center Name' in clean_final.loc[row].values])
clean_final
```

```python
# importing previous csv to see the differences
quarter_csv = "Resources_LU/LU_data_may_of_2019.csv"
three_quarters = pd.read_csv(quarter_csv)
three_quarters
```

```python
# checking to see if columns are identical

test_list1 = clean_final.columns
test_list2 = three_quarters.columns

if set(test_list1) == set(test_list2):
    print("The lists are identical")
else:
    print("The lists are not identical")
```

```python
# checking the difference between the lists

test_list1 = clean_final.columns
test_list2 = three_quarters.columns

difference_result = []
for item in test_list1:
    if item not in test_list2:
        difference_result.append(item)

print(difference_result)
```

```python
# checking for items in test list test list 2 that are not in test list 1
test_list1 = clean_final.columns
test_list2 = three_quarters.columns

difference = []
for item in test_list2:
    if item not in test_list1:
        difference.append(item)

print(difference)
```
