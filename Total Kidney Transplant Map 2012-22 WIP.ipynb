```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv('top-20-hospitals.csv')
data_df
```

```python
data_df_1 = data_df.loc[data_df["TYPE"]=="R"]
data_df_1
```

```python
data_df_2 = data_df.loc[data_df["TYPE"]=="C"]
data_df_2
```

```python
import folium
import numpy as np
from geopy.distance import geodesic
import branca

m = folium.Map([40.7, -94.5], zoom_start=4)


for index, row in data_df_1.iterrows():
    popContent = (row[0] + "<hr>" + str(row[3]))
    iframe = folium.IFrame(popContent)
    popup1 = folium.Popup(iframe,
                         min_width=200,
                         max_width=200)
    folium.CircleMarker([row[1], row[2]], popup=popup1,color='purple',
    fill_opacity=1.0,
    fill_color='purple').add_to(m)
    
for index, row in data_df_2.iterrows():
    popContent = (row[0] + "<hr>" + str(row[3]))
    iframe = folium.IFrame(popContent)
    popup1 = folium.Popup(iframe,
                         min_width=200,
                         max_width=200)
    folium.CircleMarker([row[1], row[2]], popup=popup1,color='pink',
    fill_opacity=1.0,
    fill_color='pink').add_to(m)
    
    
legend_html = '''
{% macro html(this, kwargs) %}
<div style="
    position: fixed; 
    bottom: 50px;
    left: 50px;
    width: 250px;
    height: 80px;
    z-index:9999;
    font-size:14px;
    ">
    <style>
  }
</style>
    <p><a style="color:#purple;font-size:250%;margin-left:20px;">&bull;</a>&emsp;Regular Hospital</p>
    <p><a style="color:#pink;font-size:250%;margin-left:20px;">&bull;</a>&emsp;Children's Hospital</p>
</div>
<div style="
    position: fixed; 
    bottom: 50px;
    left: 50px;
    width: 175px;
    height: 80px; 
    z-index:9998;
    font-size:14px;
    background-color: #fffff;

    opacity: 1;
    ">
</div>
{% endmacro %}
'''
legend = branca.element.MacroElement()
legend._template = branca.element.Template(legend_html)


folium.LayerControl().add_to(m)
m.get_root().add_child(legend)

m
```

```python
import folium
import numpy as np
from geopy.distance import geodesic

points = []

for index, row in data_df_1.iterrows():
    map_list = []
    map_list.append(row[1])
    map_list.append(row[2])
    points.append(map_list)


center = (np.mean([a for a,b in points]), np.mean([b for a,b in points]))
MAP = folium.Map(location = list(center), zoom_start = 5)

for p in points:
    folium.Marker(p).add_to(MAP)

folium.Marker(location = list(center),
                    radius = max([geodesic(center,point).meters for point in points])*10,
                    color = '#3186cc',
                    fill = True,
                    popup=folium.Popup("inline explicit Popup"),                   
                    fill_color = '#FF5733').add_to(MAP)
MAP
```

```python
m = folium.Map([45, 0], zoom_start=4)

folium.Marker([45, -30], popup="inline implicit popup").add_to(m)

folium.CircleMarker(
    location=[45, -10],
    radius=25,
    fill=True,
    popup=folium.Popup("inline explicit Popup"),
).add_to(m)

ls = folium.PolyLine(
    locations=[[43, 7], [43, 13], [47, 13], [47, 7], [43, 7]], color="red"
)

ls.add_child(folium.Popup("outline Popup on Polyline"))
ls.add_to(m)

gj = folium.GeoJson(
    data={"type": "Polygon", "coordinates": [[[27, 43], [33, 43], [33, 47], [27, 47]]]}
)

gj.add_child(folium.Popup("outline Popup on GeoJSON"))
gj.add_to(m)

m
```
