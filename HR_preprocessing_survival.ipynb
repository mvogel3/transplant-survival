```python
# importing dependencies

import pandas as pd
import os
import re
import calendar
```

```python
# grabbing all the files in the "Resources_HR" directory

excel_files = [file for file in os.listdir("Resources_HR") if not file.startswith("~") and file.lower().endswith(".xls")]
```

```python
# checking to see that all files were imported

print(len(excel_files))
```

```python


concat=False;
for file in excel_files:
    year_month = re.sub("[^\d]", "", file)
    year = "20"+year_month[0:2]
    month = int(year_month[2:4])
    
    
    if(year in ["2012", "2013", "2014"]):
        sheet_name = "Table 10"
    elif(year in ['2015', '2016', '2017', '2018', '2019', '2020', '2021']):
        sheet_name = "TablesC5-C10 FiguresC1-C12"
    else:
        sheet_name = "TablesC5-C12 Figures C1-C20"
        
    
    survival_df = pd.read_excel(io='Resources_HR/'+file, sheet_name=sheet_name)
#     deceased_nation_df = pd.read_excel(io="Resources_LU/"+file, sheet_name=sheet_name+" - Deceased Nation")
#     living_df = pd.read_excel(io="Resources_LU/"+file, sheet_name=sheet_name+" - Living")
#     living_nation_df = pd.read_excel(io="Resources_LU/"+file, sheet_name=sheet_name+" - Living Nation")
        
    if(len(survival_df.index)>0):
        survival_df = survival_df.drop(survival_df.index[0])
#     if(len(deceased_nation_df.index)>0):
#         deceased_nation_df = deceased_nation_df.drop(deceased_nation_df.index[0])
#     if(len(living_df.index)>0):
#         living_df = living_df.drop(living_df.index[0])
#     if(len(living_nation_df.index)>0):
#         living_nation_df = living_nation_df.drop(living_nation_df.index[0])
        
    survival_df_cleaned = pd.DataFrame()
#     living_df_cleaned = pd.DataFrame()
    
    
    columns_to_keep = ["Unnamed: 0", ' ctr_cd', 'ctr_ty', 'ENTIRE_NAME', 'CTR_CD', 'CTR_TY', 'RELEASE_DATE', 'ORG', 
                   'GSR_AD_OBS_C1M', 'GSR_AD_OBS_C1Y', 'GSR_AD_OBS_C3Y', 
                   'GSR_AD_N_C1M', 'GSR_AD_N_C1Y', 'GSR_AD_N_C3Y', 
                   'GSR_P0_OBS_C1M', 'GSR_P0_OBS_C1Y', 'GSR_P0_OBS_C3Y', 
                   'GSR_P0_N_C1M', 'GSR_P0_N_C1Y', 'GSR_P0_C3Y']
    
    for column in columns_to_keep:
        if column in survival_df:
            survival_df_cleaned[column]=survival_df[column]
        elif column+" " in survival_df:
            survival_df_cleaned[column]=survival_df[column+" "]
        elif " "+column in survival_df:
            survival_df_cleaned[column]=survival_df[" "+column]
        elif " "+column+" " in survival_df:
            survival_df_cleaned[column]=survival_df[" "+column+" "]
            
#         if column in living_df:
#             living_df_cleaned[column]=living_df[column]
#         elif column+" " in living_df:
#             living_df_cleaned[column]=living_df[column+" "]
#         elif " "+column in living_df:
#             living_df_cleaned[column]=living_df[" "+column]
#         elif " "+column+" " in living_df:
#             living_df_cleaned[column]=living_df[" "+column+" "]

#     deceased_df_cleaned["D_OR_L"] = "D"
#     deceased_nation_df["D_OR_L"] = "D"
#     living_df_cleaned["D_OR_L"] = "L"
#     living_nation_df["D_OR_L"] = "L"
    
#     center_df = pd.concat([deceased_df_cleaned, living_df_cleaned], axis=0)
#     nation_df = pd.concat([deceased_nation_df, living_nation_df], axis=0)
    
    survival_df_cleaned["YEAR"]=year
    survival_df_cleaned["MONTH"]=month
#     nation_df["YEAR"]=year
#     nation_df["MONTH"]=month
    
    if(concat==True):
        merged_survival_df= pd.concat([merged_survival_df, survival_df_cleaned], axis=0)
#         merged_nation_df= pd.concat([merged_nation_df, nation_df], axis=0)
    else:
        merged_survival_df=survival_df_cleaned
#         merged_nation_df=nation_df
        concat=True
```
