```python
# Importing dependencies

import pandas as pd
import os
import re
import calendar
import numpy as np
```

```python
KI_file = "Resources/kidney_concat.csv"
LU_file = "Resources/LU_transplant_all.csv"
HR_file = "Resources/final_center_heart.csv"
LI_file = "Resources/LI_transplant_all.csv"
```

```python
KI_df = pd.read_csv(KI_file)
KI_df
```

```python
LU_df = pd.read_csv(LU_file)
LU_df
```

```python
HR_df = pd.read_csv(HR_file)

HR_df
```

```python
LI_df = pd.read_csv(LI_file)
LI_df
```

```python
org_all = pd.concat([LI_df, LU_df, KI_df, HR_df], axis=0)
org_all
```

```python
org_drop = org_all.drop(columns=['Unnamed: 0'])

org_drop
```

```python
for col in org_drop.columns:
    if col != 'ENTIRE_NAME':
        org_drop[col].fillna(0, inplace=True)
        
org_drop
```

```python
org_all = org_drop
```

```python
org_all['RCC_TST1B_R'] = org_all['RCC_TST1B_R'] + org_all['RCC_TST1B_R']
```

```python
org_all.to_csv("Resources/transplant_sql_data.csv", index=False, header=True)
```
