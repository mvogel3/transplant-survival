```python
import pandas as pd
```

```python
# Read the Excel file with multiple tabs
excel_file = pd.ExcelFile("Resources/HR-1207.xlsx")
```

```python
# Get the list of sheet names
sheet_names = excel_file.sheet_names
```

```python
# Read data from a specific sheet by name
cenreg = excel_file.parse(sheet_names[0])
cenreg.head()
```

```python
# Read data from a specific sheet by index
nation = excel_file.parse(sheet_names[1])
nation.head()
```

```python
cenreg.columns
```

```python
HR_07 = pd.merge(cenreg, nation, how='left', on=['Unnamed: 0', ' CTR_CD', ' CTR_TY', ' RELEASE_DATE', ' ORG'])

HR_07.head()
```

```python
HR_07.rename(columns={1:'ENTIRE_NAME'})
HR_07.head()
```
