```python
# Importing dependencies

import pandas as pd
import numpy as np
import datetime as dt
import re
import calendar
```

```python
# Concatenating all the files from 2012 to 2014 since they have identical column rows

sheet_name = 'Table 7 - Deceased'
excel_files = ['LU-1207.XLS', 'LU-1301.XLS', 'LU-1307.XLS', 'LU-1401.XLS', 'LU-1406.XLS', 'LU-1412.XLS']

concat=False;
for excel in excel_files:
    df = pd.read_excel(io='Resources_LU/'+excel, sheet_name='Table 7 - Deceased')
    if(concat==True):
        concat_df = pd.concat([concat_df, df], axis=0)
    else:
        concat_df=df
        concat=True

concat_df
```

```python
# renaming the center name column to ENTIRE_NAME to match the files from 2015 and onward

concat_df.rename(columns = {"Unnamed: 0":"ENTIRE_NAME", ' RELEASE_DATE': 'RELEASE_DATE'}, inplace=True)
concat_df.head()
```

```python
# removing the column name descriptor rows from the concatenated df

df = concat_df.copy()
df = df.drop(index=[row for row in df.index if 'Comment' in df.loc[row].values])
df
```

```python
# appending a "Donor Status" column since all of these transplants were from deceased donors

df["Donor Status"] = "Deceased"
df.head()
```

```python
df.columns
```

```python
df['YEAR'].dtypes
```

```python

# df['YEAR'].astype(int)
df['YEAR'] = df.YEAR.astype('Int64')
```

```python
df.head()
```

```python
df['MONTH'] = df.MONTH.astype('Int64')
```

```python
# making COHORT_DATE column

df['COHORT_DATE'] = df['YEAR'].astype(str) + '-' + df['MONTH'].astype(str)
# df['COHORT_DATE'] = df['YEAR'].str.cat(df['MONTH'], sep = '/')
```

```python
df['COHORT_DATE'].unique()
```

```python
# editing COHORT_DATE format
# df['COHORT_DATE'] = pd.to_datetime(df['COHORT_DATE'].astype(str), format='%Y-%m')
df['COHORT_DATE'] = pd.to_datetime(df['COHORT_DATE']).dt.strftime('%Y-%m')
```

# 2017 - May 2019 (Deceased)

```python
# Concatenating all the files from 2017 onward 

sheet_names = 'Figures C1-C2 - Deceased'
files = ['LU-1701.xls', 'LU-1707.xls', 'LU-1711.xls', 'LU-1808.xls', 'LU-1811.xls', 'LU-1905.xls']

concat=False;
for file in files:
    year_month = re.sub("[^\d]", "", file)
    year = "20"+year_month[0:2]
    month = int(year_month[2:4])
    
    end_df = pd.read_excel(io='Resources_LU/'+file, sheet_name=sheet_names)
    
    end_df['YEAR']=year
    end_df['MONTH']=month
    
    if(concat==True):
        last_df = pd.concat([last_df, end_df], axis=0)
    else:
        last_df=end_df
        concat=True

last_df
```

```python
# renaming a column so it merges with the others 

last_df.rename(columns = {' RELEASE_DATE': 'RELEASE_DATE'}, inplace=True)
last_df.head()
```

```python
# appending donor status
quarter_df = last_df.copy()
quarter_df["Donor Status"] = "Deceased"
quarter_df.head()
```

```python
quarter_df['MONTH'].unique()
```

```python
# making COHORT_DATE column

quarter_df['COHORT_DATE'] = quarter_df['YEAR'].astype(str) + '-' + quarter_df['MONTH'].astype(str)
```

```python
# editing COHORT_DATE format

quarter_df['COHORT_DATE'] = pd.to_datetime(quarter_df['COHORT_DATE']).dt.strftime('%Y-%m')
quarter_df.head()
```

```python
# removing the column name descriptor rows from the concatenated df

clean_quarter = quarter_df.copy()
clean_quarter = clean_quarter.drop(index=[row for row in clean_quarter.index if 'Center Name' in clean_quarter.loc[row].values])
clean_quarter
```

```python
# checking to see if columns are identical

test_list1 = df_2016.columns
test_list2 = clean_quarter.columns

if set(test_list1) == set(test_list2):
    print("The lists are identical")
else:
    print("The lists are not identical")
```
